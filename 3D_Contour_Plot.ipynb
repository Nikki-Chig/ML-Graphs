```python
#3D Contour Plots

from mpl_toolkits import mplot3d
import numpy as np 
import matplotlib.pyplot as plt

def f(x,y):
    return np.sin(np.sqrt(x**2 + y**2))


x = np.linspace(-6,6,30)
y = np.linspace(-6,6,30)

X,Y = np.meshgrid(x,y)
Z = f(X,Y)
ax = plt.axes(projection = '3d')
ax.contour3D(X,Y,Z,50)
```
